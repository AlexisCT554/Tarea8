```python
class NodoDoble:
  def __init__(self,value,siguiente=None,previo=None):
    self.data = value
    self.next = siguiente
    self.prev = previo
```

```python
class DoubleLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None
    self.size = 0

  def get_size(self):
    return self.size
  
  def is_empty(self):
    return self.size == 0 

  def append(self,value):
    if self.is_empty():
      self.head = NodoDoble(value)
      self.tail = self.head
    else:
      self.tail.next = NodoDoble(value, None , self.tail)
      self.tail = self.tail.next
    self.size += 1

  def transversal(self):
    curr_node = self.head
    while curr_node !=None:
      print(f"| {curr_node.data} | --> " , end="")
      curr_node = curr_node.next
    print("")
  
  def remove_from_head(self,value):
    curr_node = self.head
    while curr_node.data != value:
      curr_node = curr_node.next

    if curr_node.data == value:
      print("Encontrado",curr_node.data)
      curr_node.prev.next = curr_node.next
      curr_node.next.prev = curr_node.prev
      print("", curr_node.prev.data)
    self.size -= 1
  
  def remove_from_tail(self, value):
    curr_node = self.tail
    while curr_node.data != value:
      curr_node = curr_node.prev
        
    if curr_node.data == value:
      print("Encontrado",curr_node.data) 
      curr_node.next.prev = curr_node.prev
      curr_node.prev.next = curr_node.next
    self.size -= 1
            
           
  def find_from_head(self, value):
    curr_node = self.head 
    cont = 0
    while curr_node.data != value:
      cont += 1
      curr_node = curr_node.next

    if curr_node.data == value:
      print(f"El nodo fue encontrado en la posicion {cont + 1} ingresando por head")


  def find_from_tail(self, value):
    curr_node = self.tail
    cont = 0
    while curr_node.data != value:
      cont += 1
      curr_node = curr_node.prev
        
    if curr_node.data == value:
      print(f"El nodo fue encontrado en la posicion {cont + 1} ingresando por tail")
            


  def reverse_transversal(self):
    curr_node = self.tail
    while curr_node != None:
      print(f"| {curr_node.data} | -->", end="")
      curr_node = curr_node.prev
    print("")
```

```python
ldl = DoubleLinkedList()
ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.append(40)
ldl.append(50)
ldl.append(60)

ldl.transversal()
print(f"Esta vacía? {ldl.is_empty() }")
print("\nRemove from head")
ldl.remove_from_head(30)
ldl.transversal()
print("\nRemove from tail")
ldl.remove_from_tail(50)
ldl.transversal()
print("\nFind from head")
print(f"Su Lista Doblemente Ligada tiene {ldl.get_size()} elementos")
ldl.find_from_head(20)
print("\nFind from tail")
print(f"Su Lista Doblemente Ligada tiene {ldl.get_size()} elementos")
ldl.find_from_tail(60)

print("\nCon Reverse Transversal")
ldl.reverse_transversal()
```
